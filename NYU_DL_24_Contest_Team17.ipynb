```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 512 # Reduce to 512 for faster training
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.
```

```python
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Meta-Llama-3.1-8B",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
```

## Load model and wrap with LoRA adapters

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
```

## Competition dataset

```python
# download and load competition dataset

from datasets import load_dataset
dataset = load_dataset("ad6398/nyu-dl-teach-maths-comp")
# print and see dataset
dataset
```

```python
prompt = """You are a great mathematician and you are tasked with finding if an answer to a given maths question is correct or not. Your response should be 'True' if correct, otherwise 'False'. Below is Question and Answer.


### Question:
{}


### Answer:
{}


### Solution:
{}


### Output:
{}"""


EOS_TOKEN = tokenizer.eos_token  # Must add EOS_TOKEN


def formatting_prompts_func(examples):
   questions = examples["question"]
   answers = examples["answer"]
   solutions = examples["solution"]
   outputs = examples["is_correct"]
   texts = []

   for q, a, sol, out in zip(questions, answers, solutions, outputs):
       # Must add EOS_TOKEN, otherwise your generation will go on forever!
       text = prompt.format(q, a, sol, out) + EOS_TOKEN
       texts.append(text)

   return {"text": texts}
```

```python
# Process the training dataset and generate prompt for each datapoint

train_dataset = dataset['train'].map(formatting_prompts_func, batched = True,)
```

## SFT

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

training_args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 5000,
        learning_rate = 3e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 10,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "cosine", #"linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use this for WandB etc
    )

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = train_dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 8,
    packing = False, # Can make training 5x faster for short sequences.
    args = training_args
)
```

```python
trainer_stats = trainer.train()
```

# Inference and CSV

```python
import pandas as pd
import time
import random
from collections import Counter

# Enable model in inference mode
FastLanguageModel.for_inference(model)

# Batch size is 64 as using A100
batch_size = 64

# Prepare lists to store results
ids = []
predictions = []

start_time = time.time()

# Calculate the number of batches
len_test = len(test_dataset)
num_batches = (len_test + batch_size - 1) // batch_size  # Round up for the last incomplete batch

for batch_num in range(num_batches):  # Only running one batch for demonstration
    # Gather batch examples
    batch_start = batch_num * batch_size
    batch_end = min(len_test, batch_start + batch_size)

    # Prepare batch prompts
    batch_prompts = []
    batch_ids = []

    for idx in range(batch_start, batch_end):
        # Extract question, answer, and solution for each example
        cur_ques = test_dataset['question'][idx]
        cur_ans = test_dataset['answer'][idx]
        cur_sol = test_dataset['solution'][idx]  # Include the solution for context

        # Format the input prompt to include question, answer, and solution
        input_prompt = prompt.format(
            cur_ques,  # question
            cur_ans,   # given answer
            cur_sol,   # provided solution for inference
            ""         # leave output blank for model generation
        )
        batch_prompts.append(input_prompt)
        batch_ids.append(idx)

    # Tokenize batch prompts and move to GPU
    inputs = tokenizer(batch_prompts, return_tensors="pt", padding=True, truncation=True).to("cuda")
    input_token_len = inputs['input_ids'].shape[1]

    # Run the generation process three times for each prompt
    all_responses = []
    for _ in range(3):
        # Generate model output with a maximum of 5 new tokens for more flexibility
        outputs = model.generate(**inputs, max_new_tokens=5, use_cache=True)

        # Decode batch responses
        text_responses = tokenizer.batch_decode(outputs[:, input_token_len:], skip_special_tokens=True)
        all_responses.append(text_responses)

    # Determine the most common prediction for each input
    for idx in range(len(batch_prompts)):
        responses_for_prompt = [all_responses[i][idx] for i in range(3)]

        # Extract "True" or "False" predictions from responses
        decisions = [response.split()[0] if response.split() and response.split()[0] in ["True", "False"] else None for response in responses_for_prompt]
        decisions = [decision for decision in decisions if decision is not None]  # Filter out None values

        if len(decisions) == 0:
            # If no valid "True" or "False" responses, randomly choose
            final_decision = random.choice(["True", "False"])
        elif decisions.count("True") == decisions.count("False"):
            # If there's an equal count of "True" and "False", randomly choose between them
            final_decision = random.choice(["True", "False"])
        else:
            # Otherwise, select the most common decision
            final_decision = Counter(decisions).most_common(1)[0][0]

        predictions.append(final_decision)

    # Append batch IDs
    ids.extend(batch_ids)

    # Log progress every 250 samples
    if (batch_num + 1) * batch_size % 250 == 0:
        elapsed_time = time.time() - start_time
        print(f"Processed {(batch_num + 1) * batch_size} samples in {elapsed_time:.2f} seconds")

# Create a DataFrame and save to CSV
submission_df = pd.DataFrame({
    "ID": ids,
    "is_correct": predictions
})

# Save the DataFrame as a CSV file
submission_df.to_csv("submission_sample.csv", index=False)
print("Inference complete. Results saved to 'submission_sample.csv'.")
```
